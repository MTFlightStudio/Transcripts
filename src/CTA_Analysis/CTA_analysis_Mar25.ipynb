```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.ticker import PercentFormatter
from datetime import datetime

# Set up visualization style
plt.style.use('ggplot')
sns.set(font_scale=1.2)
plt.rcParams['figure.figsize'] = (14, 8)

# Read the data
df = pd.read_csv('simplified_analytics_The Diary Of A CEO with Steven Bartlett.csv')

# Convert publish_date to datetime and remove timezone information
df['publish_date'] = pd.to_datetime(df['publish_date']).dt.tz_localize(None)

# Calculate subscribers gained per view (as a percentage)
df['subs_per_view_pct'] = (df['subscribersGained'] / df['views']) * 100

# Filter for 2025 data only
df_2025 = df[df['publish_date'].dt.year == 2025].copy()

# Create a function to assign CTA group based on publish_date
def assign_cta_group(date):
    if pd.Timestamp('2025-01-01') <= date <= pd.Timestamp('2025-01-30'):
        return "DOAC_CTA_54% (start only)"
    elif pd.Timestamp('2025-02-03') <= date <= pd.Timestamp('2025-02-17'):
        return "Coins CTA"
    elif date.strftime('%Y-%m-%d') in ['2025-02-24', '2025-02-27']:
        return "DOAC_CTA_54% (start only)"
    elif date >= pd.Timestamp('2025-03-03'):
        return "DOAC_CTA_54% (start & end)"
    else:
        return "Other"

# Apply the function to create a new column
df_2025['cta_group'] = df_2025['publish_date'].apply(assign_cta_group)

# Display sample of data with CTA groups
print("Sample of data with CTA groups:")
print(df_2025[['video_title', 'publish_date', 'cta_group', 'subs_per_view_pct']].head())

# Calculate metrics by CTA group
cta_metrics = df_2025.groupby('cta_group').agg(
    avg_subs_per_view=('subs_per_view_pct', 'mean'),
    median_subs_per_view=('subs_per_view_pct', 'median'),
    total_views=('views', 'sum'),
    total_subs=('subscribersGained', 'sum'),
    video_count=('video_id', 'count')
).reset_index()

# Calculate overall subscriber conversion rate for each CTA
cta_metrics['overall_conversion'] = (cta_metrics['total_subs'] / cta_metrics['total_views']) * 100

# Display CTA group metrics
print("\nCTA Group Performance Metrics:")
print(cta_metrics[['cta_group', 'avg_subs_per_view', 'median_subs_per_view', 
                    'overall_conversion', 'video_count', 'total_views', 'total_subs']])

# Create bar chart comparing CTA effectiveness
plt.figure(figsize=(14, 8))
ax = sns.barplot(x='cta_group', y='avg_subs_per_view', data=cta_metrics, palette='viridis')
ax.set_title('Average Subscriber Conversion Rate by CTA Strategy', fontsize=16)
ax.set_xlabel('CTA Strategy', fontsize=14)
ax.set_ylabel('Average Subscribers per View (%)', fontsize=14)

# Add video count annotations
for i, row in enumerate(cta_metrics.itertuples()):
    ax.text(i, row.avg_subs_per_view + 0.05, f'Videos: {row.video_count}\nTotal views: {row.total_views:,}', 
            ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()

# Create box plot to show distribution within each CTA group
plt.figure(figsize=(14, 8))
sns.boxplot(x='cta_group', y='subs_per_view_pct', data=df_2025, palette='viridis')
plt.title('Distribution of Subscriber Conversion Rates by CTA Strategy', fontsize=16)
plt.xlabel('CTA Strategy', fontsize=14)
plt.ylabel('Subscribers per View (%)', fontsize=14)
plt.tight_layout()
plt.show()

# Create time series visualization showing each video with CTA group
plt.figure(figsize=(16, 10))
for cta, color in zip(sorted(df_2025['cta_group'].unique()), 
                    ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']):
    group_data = df_2025[df_2025['cta_group'] == cta]
    plt.scatter(group_data['publish_date'], group_data['subs_per_view_pct'], 
               label=cta, s=100, alpha=0.7, color=color)

plt.title('Subscriber Conversion Rate by Publish Date and CTA Strategy', fontsize=16)
plt.xlabel('Publish Date', fontsize=14)
plt.ylabel('Subscribers per View (%)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)

# Add lines showing CTA period transitions
transition_dates = ['2025-01-30', '2025-02-03', '2025-02-17', '2025-02-24', '2025-02-27', '2025-03-03']
for date in transition_dates:
    plt.axvline(x=pd.to_datetime(date), color='gray', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# Show top performing videos in each CTA group
print("\nTop performing videos in each CTA group:")
for cta in df_2025['cta_group'].unique():
    group_data = df_2025[df_2025['cta_group'] == cta].sort_values('subs_per_view_pct', ascending=False)
    print(f"\n{cta} - Top 3 Videos:")
    for i, row in group_data.head(3).iterrows():
        print(f"- {row['video_title']} ({row['subs_per_view_pct']:.2f}%) - {row['publish_date'].strftime('%Y-%m-%d')}")
```
